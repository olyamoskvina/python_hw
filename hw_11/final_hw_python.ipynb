```python
f(x) = -5*x**2 + 10*x - 150
Определить корни
Найти интервалы, на которых функция возрастает
Найти интервалы, на которых функция убывает
Построить график
Вычислить вершину
Определить промежутки, на котором f > 0
Определить промежутки, на котором f < 0
```

```python
import math
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
fx = -5*x**2 + 10*x - 150
fx
```

```python
p = plot (fx) # график
```

```python
p = plot(fx, (x, -15, 15)) # график с интервалом
```

```python
sqrt_1, sqrt_2  = solve(fx) # корни
print(sqrt_1, sqrt_2)
```

```python
pr = diff(fx)
print(pr)
```

```python
apex = solve(diff(fx)) # вершина
print(apex)
```

```python
solve(diff(fx) > 0) #интервалы, на который функция возрастает
```

```python
solve(diff(fx) < 0) #интервалы, на который функция убывает
```

```python
solve(fx > 0)  # промежутки, на котором f > 0
```

```python
solve(fx < 0) # промежутки, на котором f < 0
```
